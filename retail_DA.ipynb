# find the higest revenue generating product

```python
query=""" select product_id,sum(sale_price) as sales 
    from orders_m 
    group by product_id
    order by sales desc
    limit 10"""

cur.execute(query)

data=cur.fetchall()

df = pd.DataFrame(data, columns=[col[0] for col in cur.description])
df
```

# find top 5 highest selling products in each region

```python
query=""" select region,product_id,sum(sale_price) as sales 
    from orders_m 
    group by region,product_id
    order by region,sales desc"""

cur.execute(query)

data=cur.fetchall()

df = pd.DataFrame(data, columns=[col[0] for col in cur.description])
df
```

```python
query=""" with cte as(select region,product_id,sum(sale_price) as sales 
    from orders_m 
    group by region,product_id)
    select * from( select * ,
    row_number() over (partition by region order by sales desc) as rn
    from cte) A
    where rn<=5"""

cur.execute(query)

data=cur.fetchall()

df = pd.DataFrame(data, columns=[col[0] for col in cur.description])
df
```

# find month over month growth comparison for 2022 and 2023 sales eg: jan 2022 vs jan 2023

```python
query=""" with cte as(select year(order_date) as order_year,
            month(order_date) as order_month ,
            sum(sale_price) as sales
        from orders_m
        group by  year(order_date),month(order_date)) 
       select order_month,
           sum(case when order_year=2022 then sales else 0 end) as sales_2022,
           sum(case when order_year=2023 then sales else 0 end) as sales_2023
        from cte
        group by order_month
        order by order_month"""
cur.execute(query)

data=cur.fetchall()

df = pd.DataFrame(data, columns=[col[0] for col in cur.description])
df
```

# for each category which month has highest sales

```python
query=""" with cte as (select category,
        format(order_date,'yyyyMM') as month,
        sum(sale_price) as total_sales
    from orders_m 
    group by category, format(order_date,'yyyyMM'))
    select * from(
    select *,
    row_number() over (partition by category order by total_sales desc)as rn
    from cte) a
    where rn=1
    """

cur.execute(query)

data=cur.fetchall()

df = pd.DataFrame(data, columns=[col[0] for col in cur.description])
df
```

# which subcategory has highest growth by profit in 2023 compate to 2022

```python
query="""with cte as(select sub_category,year(order_date) as order_year,
            sum(sale_price) as sales
        from orders_m
        group by  year(order_date),sub_category) 
       ,cte2 as 
       (select sub_category,
           sum(case when order_year=2022 then sales else 0 end) as sales_2022,
           sum(case when order_year=2023 then sales else 0 end) as sales_2023
        from cte
        group by sub_category)
        select * ,
        (sales_2023-sales_2022)*100/sales_2022 as Growth
        from cte2
        order by growth desc
        limit 1
    """

cur.execute(query)

data=cur.fetchall()

df = pd.DataFrame(data, columns=[col[0] for col in cur.description])
df
```
